```python
def is_valid_mobile(phone_number):
    cleaned_number = ''.join(filter(str.isdigit, phone_number))

    if cleaned_number.startswith('91') or cleaned_number.startswith('+91'):
        return 6000000000 <= int(cleaned_number[3:]) <= 9999999999
    return False

for entry in data1:
    entry['isValidMobile'] = is_valid_mobile(entry['phoneNumber'])

for entry in data1:
    print(entry)
```

```python
import json
import hashlib
from datetime import datetime
import pandas as pd

with open('/content/DataEngineeringQ2.json', 'r') as file:
    data = json.load(file)

def validate_and_hash_phone(number):
    if number is None:
        return False, None
    valid_prefixes = ('+91', '91')
    if number.startswith(valid_prefixes) and len(number) in (12, 13) and number[-10:].isdigit() and 6000000000 <= int(number[-10:]) <= 9999999999:
        return True, hashlib.sha256(number[-10:].encode()).hexdigest()
    else:
        return False, None


def extract_date(dob):
    if dob:
        birth_date = datetime.strptime(dob, '%Y-%m-%dT%H:%M:%S.%fZ')
        return birth_date.strftime('%Y-%m-%d')
    else:
        return None


columns = ['appointmentId', 'fullName', 'phoneNumber', 'isValidMobile', 'phoneNumberHash', 'gender', 'DOB', 'Age', 'noOfMedicines', 'noOfActiveMedicines', 'noOfInActiveMedicines', 'MedicineNames']
df_data = []

for record in data:
    patient = record['patientDetails']
    consultation = record['consultationData']

    gender = patient.get('gender', 'others').upper()
    gender = 'male' if gender == 'M' else 'female' if gender == 'F' else 'others'

    dob = patient.get('birthDate')
    age = calculate_age(dob)
    dob = extract_date(dob)

    fullName = f"{patient.get('firstName', '')} {patient.get('lastName', '')}".strip()

    phoneNumber = patient.get('phoneNumber')
    is_valid, phone_hash = validate_and_hash_phone(phoneNumber)

    medicine_names = [med['name'] for med in consultation['medicines'] if 'name' in med]
    no_of_medicines = len(medicine_names)
    active_medicines = [med for med in consultation['medicines'] if med.get('IsActive', False)]
    no_of_active_medicines = len(active_medicines)
    no_of_inactive_medicines = no_of_medicines - no_of_active_medicines


    df_data.append([
        record['appointmentId'],
        fullName,
        phoneNumber,
        is_valid,
        phone_hash,
        gender,
        dob,
        age,
        no_of_medicines,
        no_of_active_medicines,
        no_of_inactive_medicines,
        '~'.join(medicine_names)
    ])

df = pd.DataFrame(df_data, columns=columns)

df.to_csv('/content/ProcessedData.csv', index=False, sep='~')
```
